```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
# Criar um espa√ßo vetorial
x = np.linspace(0, 5, 11)
x
```

```python
y = x*x
y
```

```python
# Forma Funcional
plt.plot(x, y, color='r')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Titulo')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x,y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(y,x,'g*-')
```

```python
# Cria Caixa Vazia
fig = plt.figure()

# Cria um eixo
axes1 = fig.add_axes([0.1,0.1, 0.8,0.8])
axes2 = fig.add_axes([0.2,0.3, 0.3,0.3])

axes1.plot(x,y)
axes1.set_xlabel('Eixo X')
axes1.set_title('Titulo')
axes2.plot(y,x)
```
